```python
total_gene_count = pd.Series(dtype='int64')

def count_interactions(file, col = 0):
    global total_gene_count
    df = pd.read_csv("patient_networks/CD_97_5_all_interactions/"+file, sep='\t',header=None)
    gene_count = df.value_counts(subset=[col])
    try:
        total_gene_count = total_gene_count.add(gene_count, fill_value =0)
    except:
        total_gene_count = total_gene_count.append(gene_count)
    return total_gene_count

def main():
    directory = 'patient_networks/CD_97_5_all_interactions/'
    files = os.listdir(directory)
    for file in files:
        total_col_0 = count_interactions(file)
        total_col_1 = count_interactions(file, col=1)
        try:
            total = total_col_0.add(total_col_1, fill_value=0)
        except:
            total = total_col_0.append(total_col_1)
    return(total)

CD_97_5_all_node_degree = main()
```
