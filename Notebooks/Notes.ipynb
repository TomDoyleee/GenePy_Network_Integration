```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import os


# change wd
os.chdir("/Users/tomdoyle/Documents/University/Southampton/Course/BIOL6068-Research_Project/Python")

# genepy matrix file in relation to wd
genepy_matrix_file = "../Data/IBD_genepy1_4_2022.matrix"


genepy_df = pd.read_table(genepy_matrix_file, index_col = 'Samid')
    
# normalise genepy across genes using Min max scale, every value to be between 0 and 1.
def norm_genepy():
    '''
    Function to normalise across genes for each patient
    '''
    genepy_normalized_hold = pd.DataFrame()
    
    for i in range(len(genepy_df.columns)):
        col = preprocessing.minmax_scale(genepy_df.iloc[:,i])
        col_series = pd.Series(col,
                               index = genepy_df.index,
                               name=genepy_df.columns[i]))
        # adds each line to the data frame 
        genepy_normalized_hold = genepy_normalized_hold.append(col_series, 
                                                               ignore_index=True)
    #adds index values (patientID) back
    genepy_normalized_hold.index = genepy_df.index
    
    return genepy_normalized_hold
    
genepy_normalized = norm_genepy()
genepy_normalized.head()
```
